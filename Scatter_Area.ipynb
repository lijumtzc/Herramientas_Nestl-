```python
import pandas as pd
import matplotlib
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
import plotly.express as px    
%matplotlib inline
```

```python
df = pd.read_excel('Libro2.xlsx', index_col=None)
print(df.head())
```

```python
df['X'] = df['Mix'].astype(float)
df['Y'] = df['SOG'].astype(float)

datos_ddi = df['X']
datos_ventas = df['Y']
s = df['Color']
```

```python
data_cuenta = df.Color.unique()
cuenta = data_cuenta.size
print(data_cuenta)
print("La cuenta es de: ",cuenta)
```

```python
data_frame_lugar = df['Color']
lista_area1 = []
lista_area2 = []
lista_area3 = []
lista_area4 = []
lista_total = []
count_general = 0
count_area1 = 0
count_area2 = 0
count_area3 = 0
count_area4 = 0
media_x = datos_ddi.mean()
media_y = datos_ventas.mean()
serie_x = datos_ddi
serie_y = datos_ventas

for x in serie_x:
    #print(data_frame_lugar[count_general] + " " +str(x) +" " +str(serie_y[count_general]) )
    if x < media_x and serie_y[count_general]> media_y:
        count_area1 = count_area1 +1 
        lista_area1.append(data_frame_lugar[count_general])
    if x > media_x and serie_y[count_general]> media_y:
        count_area2 = count_area2 +1
        lista_area2.append(data_frame_lugar[count_general])
    if x < media_x and serie_y[count_general]< media_y:
        count_area3 = count_area3 +1
        lista_area3.append(data_frame_lugar[count_general])
    if x > media_x and serie_y[count_general]< media_y:
        count_area4 = count_area4 +1
        lista_area4.append(data_frame_lugar[count_general])
    count_general = count_general + 1
    
lista_total.append((count_area1,count_area2,count_area3,count_area4))

print("Area1: ",count_area1)
print("Area2: ",count_area2)
print("Area3: ",count_area3)
print("Area4: ",count_area4)
```

```python
df0 = pd.DataFrame(data=["Area 1", "Area 2", "Area 3", "Area 4"],columns=['Areas'])
df_total_area = pd.DataFrame(data = lista_total)
df_total_area = df_total_area.T
df_total_area = df_total_area.rename(columns = {0: "Total"})
df_total_por_areas = pd.concat([df0, df_total_area], axis = 1)
df_total_por_areas
```
