# **Acumuladas** 
---
Programa para calcular la distribución de frecuencias acumuladas.

```python
# Librerías
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
from scipy.stats import mannwhitneyu
from scipy import stats
%matplotlib inline 
```

```python
df = pd.read_excel('/home/lidia/Documentos/Sep15/Frecuencias.xlsx')
print(df.columns)
print(df.shape)
```

```python
df_q1 = df.query('Ventas < Ventas.quantile(.25)')
df_q2 = df.query('Ventas > Ventas.quantile(.25) & Ventas < Ventas.quantile(.50)')
df_q3 = df.query('Ventas > Ventas.quantile(.50) & Ventas < Ventas.quantile(.75)')
df_q4 = df.query('Ventas > Ventas.quantile(.75)')
```

```python
# Parámetros estéticos de seaborn
sns.set_palette("gnuplot", desat = 1)
sns.set_context(rc={"figure.figsize": (10, 10)})

fig = plt.figure(figsize=(8,8))

fig.add_subplot(2, 2, 1)
#**************************************************
Ax = np.sort(df_q1['Ventas'])
s = 1/len(Ax)
Ay = [s*i for i in range(0,len(Ax))]
plt.plot(Ax,Ay, color = '#2A6FDB', alpha = 0.8, linewidth = 2)
plt.grid(True);
#***************************************************
ndat, minmax, mean, var, skew, kurt = stats.describe(df_q1['Ventas'])
std = np.std(df_q1['Ventas'])
textstr = '$\overline{x} =%.4f$ \n$s=%.2f$ \n$\mathit{skew}= %.2f$ \n$\mathit{kurt}= %.2f$' % (mean,std,skew,kurt)
props=dict(boxstyle='round',facecolor='#2A6FDB',alpha=0.4)
plt.text(1000,0,textstr,color='#191970',bbox=props)
#******************************************************************************
fig.add_subplot(2, 2, 2)
#***************************************************
Axc = np.sort(df_q2['Ventas'])
sc = 1/len(Axc)
Ayc = [sc*i for i in range(0,len(Axc))]
plt.plot(Axc,Ayc, color = '#FB7756', alpha = 1, linewidth = 2)
plt.grid(True);
#***************************************************
ndat, minmax, mean, var, skew, kurt = stats.describe(df_q2['Ventas'])
std = np.std(df_q2['Ventas'])
textstr = '$\overline{x} =%.4f$ \n$s=%.2f$ \n$\mathit{skew}= %.2f$ \n$\mathit{kurt}= %.2f$' % (mean,std,skew,kurt)
props=dict(boxstyle='round',facecolor='#FB7756',alpha=0.4)
plt.text(3900,0,textstr,color='#191970',bbox=props)
#******************************************************************************
fig.add_subplot(2, 2, 3)
#***************************************************
Ax = np.sort(df_q3['Ventas'])
s = 1/len(Ax)
Ay = [s*i for i in range(0,len(Ax))]
plt.plot(Ax,Ay, color = '#1AC0C6', alpha = 1, linewidth = 2)
plt.grid(True);
#***************************************************
ndat, minmax, mean, var, skew, kurt = stats.describe(df_q3['Ventas'])
std = np.std(df_q3['Ventas'])
textstr = '$\overline{x}=%.4f$ \n$s=%.2f$ \n$\mathit{skew}= %.2f$ \n$\mathit{kurt}= %.2f$' % (mean,std,skew,kurt)
props=dict(boxstyle='round',facecolor='#1AC0C6',alpha=0.4)
plt.text(9400,0,textstr,color='#191970',bbox=props)
#******************************************************************************
fig.add_subplot(2, 2, 4)
#***************************************************
Axc = np.sort(df_q4['Ventas'])
sc = 1/len(Axc)
Ayc = [sc*i for i in range(0,len(Axc))]
plt.plot(Axc,Ayc, color = '#836BB3', alpha = 1, linewidth = 2)
plt.grid(True);
#***************************************************
ndat, minmax, mean, var, skew, kurt = stats.describe(df_q4['Ventas'])
std = np.std(df_q4['Ventas'])
textstr = '$\overline{x}=%.4f$ \n$s=%.2f$ \n$\mathit{skew}= %.2f$ \n$\mathit{kurt}= %.2f$' % (mean,std,skew,kurt)
props=dict(boxstyle='round',facecolor='#836BB3',alpha=0.2)
plt.text(79000,0,textstr,color='#191970',bbox=props)
#******************************************************************************

angle = 90

fig.text(0.07,0.5, r'$Distribución\ de\ Probabilidad\ Acumulada$',ha='center',va='center',fontsize='16',rotation=angle, rotation_mode='anchor')
fig.text(0.5,0.92, r'Cliente4 Dsc Nestle',ha='center',va='center',fontsize='20')
fig.text(0.5,0.075, r'$Valor$',ha='center',va='center',fontsize='16')

#plt.savefig("EGOA.pdf")
plt.savefig("Acumladas_Frecuencias.jpg")
```
